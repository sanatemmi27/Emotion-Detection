### BUILDING THE CNN MODEL

**Once we extracted the data from the HDF5 file and saved in our local directory, we can now start building the CNN model.**

```python
#import all the required libraries for building the CNN model
import keras
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dropout, BatchNormalization
from keras.layers import Flatten, Activation, Dense
from keras.layers.advanced_activations import ELU
from keras import optimizers
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
from keras import backend as K
from keras.callbacks import TensorBoard, ModelCheckpoint

import time
```

```python
train_dir = 'F:/desktop projects/PRACTICE/notebooks/Completed Projects/Emotion Detection/fer2013/hdf5/train'
val_dir = 'F:/desktop projects/PRACTICE/notebooks/Completed Projects/Emotion Detection/fer2013/hdf5/val'
```

```python
# augmentation, then initialize the image preprocessor
train_datagen = ImageDataGenerator(rotation_range = 10, 
                              zoom_range = 0.1, 
                              horizontal_flip = True, 
                              rescale = 1 / 255., 
                              fill_mode = "nearest",
                             height_shift_range = 0.1,
                             width_shift_range = 0.1)

val_datagen = ImageDataGenerator(rescale = 1 / 255.)

train_data = train_datagen.flow_from_directory(train_dir, target_size = (48, 48), batch_size = 32)
val_data = val_datagen.flow_from_directory(val_dir, target_size = (48, 48), batch_size = 32)
```

```python
#Building the convolutional neural network
dense_layers = [0, 1, 2]
layer_sizes = [32, 64, 128]
conv_layers = [1, 2, 3]
dropout_conv = [0.2, 0.3]

for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            for dropout in dropout_conv:
                NAME = "{}-dropout---{}-conv---{}-nodes---{}-dense---{}".format(dropout, conv_layer, layer_size, dense_layer, int(time.time()))
                print(NAME)

                model = Sequential()
                model.add(Conv2D(layer_size, (3, 3), activation = 'relu', input_shape = (48, 48, 3), padding = 'same'))
                model.add(BatchNormalization())
                model.add(MaxPooling2D((2, 2)))
                model.add(Dropout(dropout))
                
                model.add(Conv2D(layer_size, (3, 3), activation = 'relu', input_shape = (48, 48, 3), padding = 'same'))
                model.add(BatchNormalization())
                model.add(MaxPooling2D((2, 2)))
                model.add(Dropout(dropout))

                for l in range(conv_layer-1):
                    model.add(Conv2D(layer_size, (3, 3), activation = 'relu', padding = 'same'))
                    model.add(BatchNormalization())
                    model.add(MaxPooling2D((2, 2)))
                    model.add(Dropout(dropout))

                model.add(Flatten())

                for _ in range(dense_layer):
                    model.add(Dense(layer_size))
                    model.add(Activation('relu'))
                    model.add(BatchNormalization())
                    model.add(Dropout(0.5))

                model.add(Dense(6, activation = 'softmax'))

                tensorboard = TensorBoard(log_dir = "logs1/{}".format(NAME))

                #Compile the model
                model.compile(loss = 'categorical_crossentropy', optimizer = optimizers.Adam(lr = 0.001), metrics = ['accuracy'])
                
#                 filepath = "F:/weight file/weights.best.hdf5"
#                 checkpoint = ModelCheckpoint(filepath, monitor = 'val_acc', verbose = 1, save_best_only = True, mode = 'max')

                history = model.fit_generator(train_data, steps_per_epoch = 100, epochs = 50,
                                                  validation_data = val_data, validation_steps = 50,
                                                  callbacks = [tensorboard])
```

**One we are done with the above permutations and combination of the hyperparameters, we will choose the hyperparameters corresponding to the best performing model and will construct a new CNN architecture with choosen hyperparameters and train the model for larger epochs to obtain the best accuracy.**

```python
model1 = Sequential()
model1.add(Conv2D(128, (3, 3), activation = 'relu', input_shape = (48, 48, 3), padding = 'same'))
model1.add(BatchNormalization())
model1.add(MaxPooling2D((2, 2)))
model1.add(Dropout(0.2))
                
model1.add(Conv2D(128, (3, 3), activation = 'relu', padding = 'same'))
model1.add(BatchNormalization())
model1.add(MaxPooling2D((2, 2)))
model1.add(Dropout(0.2))

model1.add(Conv2D(128, (3, 3), activation = 'relu', padding = 'same'))
model1.add(BatchNormalization())
model1.add(MaxPooling2D((2, 2)))
model1.add(Dropout(0.2))

model1.add(Flatten())

# model.add(Dense(layer_size))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))

model1.add(Dense(6, activation = 'softmax'))

#tensorboard = TensorBoard(log_dir = "logs/{}".format(NAME))

#Compile the model
model1.compile(loss = 'categorical_crossentropy', optimizer = optimizers.Adam(lr = 0.001), metrics = ['accuracy'])
                
filepath = "F:/desktop projects/PRACTICE/notebooks/Completed Projects/Emotion Detection/models/weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor = 'val_acc', verbose = 1, save_best_only = True, mode = 'max')

history1 = model1.fit_generator(train_data, steps_per_epoch = 100, epochs = 300,
                              validation_data = val_data, validation_steps = 50,
                              callbacks = [checkpoint])
```
